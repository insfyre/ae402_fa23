```python
#Q2
mu = 398600 #km**3/s**2
#a)
a=8000
e = 0.125
i = np.deg2rad(10)
omega = np.deg2rad(45) #RAAN, right angle of ascending node
w = np.deg2rad(10) #argument of periapsis
M = np.deg2rad(170) #Mean anomaly

#Find eccentric anomaly E using M and e
#I love newton's method!
E0 = M
g = 1
itr = 0 #iteration counter
tol = 1e-10
print("Newton's Method:")
while(abs(g) > tol): #see Lecture 4 page 18
# for i in range(4):
    g = E0-e*np.sin(E0) - M
    dg_dE = 1-e*np.cos(E0) #derivative dg over dE
    E1 = E0 - g/dg_dE
    print(itr,g,E1)
    E0 = E1 #update E0 as E1
    itr = itr+1
print(f"Q2.1 Eccentric Anomaly after Newton's Method: {E1}")

#true anomaly
#lecture 4 page 18
f = 2*np.arctan2( np.sqrt(1+e)*np.tan(E0/2), np.sqrt(1-e))
print(f"Q2.2 f = {f}")

#angular momentum known a, e, mu
h = np.sqrt(mu*a*(1-e**2))
print(f"Q2.3 h = {h}")

#theta
theta = w + f
print(f"Q2.4 theta = {theta}")

#r_mag
r_mag = a*(1-e**2) / (1+e*np.cos(f))
print(f"Q2.5 r_mag = {r_mag}")

#r_vec
#lecture 5 page 9
e_hat_r = np.array([np.cos(theta)*np.cos(omega) - np.cos(i)*np.sin(omega)*np.sin(theta), 
                    np.cos(theta)*np.sin(omega) + np.cos(i)*np.cos(omega)*np.sin(theta), 
                    np.sin(i)*np.sin(theta)])
r = r_mag*e_hat_r
print(f"Q2.6 r = {repr(r)}")

#v
v = np.array([
    -mu/h*( np.cos(omega)*(np.sin(theta) + e*np.sin(w)) + np.sin(omega)*(np.cos(theta) + e*np.cos(w))*np.cos(i) ),
    -mu/h*( np.sin(omega)*(np.sin(theta) + e*np.sin(w)) - np.cos(omega)*(np.cos(theta) + e*np.cos(w))*np.cos(i) ),
    mu/h*( (np.cos(theta) + e*np.cos(w))*np.sin(i))
])
print(f"Q2.7 v = {repr(v)}")
```

## Question 3: Phasing Tiem

```python
#Q3
#I love phasing!

#knowns/givens:
mu = 398600 #km**3/s**2
rE = 6378 #km
a = 26553 #km
e = 0.0 #circular orbit
i = 55 #degrees!
deltaT = 15*60/3 #divided by 3 since it's over 3 orbits

# #referencing Lecture 10 example 1:
# E_b = deltaT/np.sqrt(a**3/mu)
P = 2*np.pi*np.sqrt(a**3/mu)
P_phasing = P-deltaT #seconds
print(f"Period of phasing: {P_phasing}")

a_phasing = (mu*(P_phasing/(2*np.pi))**2)**(1/3)
print(f"semimajor axis of phasing: {a_phasing}")

#3.3 transfer time
transfer_time = P_phasing*3/3600 #hours
print(f"total transfer time: {transfer_time}")

#3.4 velocity on original orbit
v = np.sqrt(mu/a) #circular orbit
print(f"v_original = {v}")
#3.5 r_phasing_a apogee
r_phasing_a = a #service spacecraft needs to go into a smaller orbit to catch up
print(f"r_phasing_a = {r_phasing_a}")

#3.6 r_phasing_p perigee
r_phasing_p = 2*a_phasing - r_phasing_a
print(f"r_phasing_p = {r_phasing_p}")

#3.7 eccentricity
e_phasing = (r_phasing_a - r_phasing_p)/(r_phasing_a + r_phasing_p)
print(f"e_phasing = {e_phasing}")

#3.8 angular momentum
h_phasing = np.sqrt(a_phasing*mu*(1-e_phasing**2))
print(f"h_phasing = {h_phasing}")

#3.9 v_phasing_a
v_phasing_a = h_phasing/r_phasing_a
print(f"v_phasing_a = {v_phasing_a}")

# v_phasing_p = h_phasing/r_phasing_p 

delta_v_Adep = v_phasing_a - v #use apogee or perigee?? (use apogee since burns take place at apogee (a_original)
delta_v_Aarr = v - v_phasing_a
delta_v = abs(delta_v_Adep)+abs(delta_v_Aarr)
print(f"Total Delta V needed to change phase: {delta_v} km/s")
# print(v_phasing_p)
```

# Q4: Comparing Hohmann vs Bi-elliptic

```python
#Q4
#follow solution in Lecture 9

r1 = 7000
r2 = 105000
rb = 210000
mu = 398600

#hohmann
aH = (r1+r2)/2
print(f"Hohmann semimajor axis: {aH}")

vt_dep = np.sqrt(mu*(2/r1 - 1/aH))
print(f"Hohmann periapsis v: {vt_dep}")

vt_arr = np.sqrt(mu*(2/r2 - 1/aH))
print(f"Hohmann apoapsis v: {vt_arr}")

v1 = np.sqrt(mu/r1)
print(f"Q4.4 v1 = {v1}")

v2 = np.sqrt(mu/r2)
print(f"Q4.5 v2 = {v2}")

dvA = vt_dep - v1
dvB = v2 - vt_arr
dV = abs(dvA) + abs(dvB)
print(f"Q4.6 Hohmann Delta V: {dV}")

#Bielliptic
aH2 = (r1+rb)/2
print(f"Q4.7 semimajor H2 {aH2}")

vA2 = np.sqrt(mu*(2/r1 - 1/aH2))
print(f"Q4.8 velocity perigee leg 1 {vA2}")

dv1 = abs(vA2-v1)
print(f"Q4.9 dv1 {dv1}")

vB2 = np.sqrt(mu*(2/rb - 1/aH2))
print(f"Q4.10 velocity apogee leg 1 {vB2}")

aH3 = (rb + r2)/2
print(f"Q4.11 semimajor H3 {aH3}")

vB3 = np.sqrt(mu*(2/rb - 1/aH3))
print(f"Q4.12 velocity apogee leg 2 {vB3}")

dv2 = abs(vB3 - vB2)
print(f"Q4.13 dv2 {dv2}")

vC3 = np.sqrt(mu*(2/r2 - 1/aH3))
print(f"Q4.14 velocity perigee leg 2 {vC3}")

dv3 = abs(v2 - vC3)
print(f"Q4.15 dv3 {dv3}")

dVt = dv1 + dv2 + dv3
print(f"Q4.16 Bielliptic Delta V: {dVt}")

print("Need to explore if bielliptic is ever higher than Hohmann")
rb_arr = np.linspace(r2, 940000, 500)
dV_arr = []
for r_b in rb_arr:
    aH2_temp = (r1+r_b)/2
    vA2_temp = np.sqrt(mu*(2/r1 - 1/aH2_temp))
    dv1_temp = abs(vA2_temp-v1)
    vB2_temp = np.sqrt(mu*(2/r_b - 1/aH2_temp))
    aH3_temp = (r_b + r2)/2
    vB3_temp = np.sqrt(mu*(2/r_b - 1/aH3_temp))
    dv2_temp = abs(vB3_temp - vB2_temp)
    vC3_temp = np.sqrt(mu*(2/r2 - 1/aH3_temp))
    dv3_temp = abs(v2 - vC3_temp)
    dV_temp = dv1_temp + dv2_temp + dv3_temp
    dV_arr.append(dV_temp)
plt.plot(rb_arr,dV_arr)
plt.plot(rb_arr,dV*np.ones_like(rb_arr))
plt.show()

r_b = rb
dV_temp = dVt
while dV_temp < dV:
    aH2_temp = (r1+r_b)/2
    vA2_temp = np.sqrt(mu*(2/r1 - 1/aH2_temp))
    dv1_temp = abs(vA2_temp-v1)
    vB2_temp = np.sqrt(mu*(2/r_b - 1/aH2_temp))
    aH3_temp = (r_b + r2)/2
    vB3_temp = np.sqrt(mu*(2/r_b - 1/aH3_temp))
    dv2_temp = abs(vB3_temp - vB2_temp)
    vC3_temp = np.sqrt(mu*(2/r2 - 1/aH3_temp))
    dv3_temp = abs(v2 - vC3_temp)
    dV_temp = dv1_temp + dv2_temp + dv3_temp
    # print(dV_temp)
    r_bfin = r_b
    r_b = r_b-1  #iterate lower
print(f"dV = {dV_temp}")
print(f"rB = {r_bfin}")
r_b += 2
aH2_temp = (r1+r_b)/2
vA2_temp = np.sqrt(mu*(2/r1 - 1/aH2_temp))
dv1_temp = abs(vA2_temp-v1)
vB2_temp = np.sqrt(mu*(2/r_b - 1/aH2_temp))
aH3_temp = (r_b + r2)/2
vB3_temp = np.sqrt(mu*(2/r_b - 1/aH3_temp))
dv2_temp = abs(vB3_temp - vB2_temp)
vC3_temp = np.sqrt(mu*(2/r2 - 1/aH3_temp))
dv3_temp = abs(v2 - vC3_temp)
dV_temp = dv1_temp + dv2_temp + dv3_temp
print(f"rB plus 1: {r_b} dV(rB+1): {dV_temp}")
# #knowns/givens:
# alt2 = 500
# alt1 = 200
# rE = 6378 #km
# r2 = rE + alt2
# r1 = rE + alt1
# aH = (r1+r2)/2
# mu = 398600
# vP = 8 #km/s

# dv2 = np.sqrt(mu*(2/r2 - 1/aH)) - np.sqrt(mu/r2)
# # dv1 = np.sqrt(mu/r1) - np.sqrt(mu*(2/r1 - 1/aH))
# dV = dv2 #assuming we have single burn because we just need a minimum altitude of 200km
# #problem didn't say shit about circularizing
# print(f"Delta V:{dV} km/s")

# T = np.pi*np.sqrt(aH**3/mu) #time spent in transfer orbit, half of regular period calc
# print(f"Period: {T} s  =  {T/60} min")
# print("Looks correct, ISS orbits in 90 min at 400km altitude")
# # dv2
```
